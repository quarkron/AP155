## Euler's Method

$$x(t+h) = x(t) + hf(x,t)$$
solves:
$$ \frac{\text{d}x}{\text{d} t} = f(x,t)$$

Solve the differential equation: 
$$ \frac{\text{d}x}{\text{d} t} = -x^3 + \sin t $$
with initial condition $x(0) = 0$

```python
from math import sin
import numpy as np
import matplotlib.pyplot as plt
def em(f, x0, a, b, N=1000):
    ''' Solves an ODE using Euler Method
    f = function in x,t 
    x0 = initial condition
    a = starting point
    b = endpoint 
    N = number of steps
    '''
    h = (b-a)/N 
    ts = np.linspace(a,b,N)
    xs = [x0]
    for t in np.arange(0,N-1,1): 
        xs.append(xs[t]+h*f(xs[t],ts[t]))
    plt.figure(dpi=150)
    plt.plot(ts, xs)

em(f, 0, 0, 10)
```
