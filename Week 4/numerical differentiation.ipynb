```python
import numpy as np
import matplotlib.pyplot as plt
```

Problem: Create a plot of the analytical and numerical derivative (using linear central difference) of 
$$f(x) = 1 + \frac{1}{2}\tanh 2x$$

```python
def f(x):
    return (0.5)*np.tanh(2*x)+1 #use np.tanh not math.tanh

def g(x):
    return 1/((np.cosh(2*x))**2)

def central_diff(f, x, h=0.01):
    cd = (f(x+h/2) - f(x-h/2))/h
    return cd

def forward_diff(f, x, h=0.01):
    return (f(x) - f(x-h))/h 

def back_diff(f,x, h=0.01):
    return (f(x+h) - f(h))/h
```

```python
a,b = -2,2
step = 0.01
x_vals = np.arange(a,b+step, step)
y_vals = f(x_vals)
num_derivative = central_diff(f, x_vals, h=0.1)

plt.figure(dpi=150)
# plt.plot(x_vals, y_vals, label='function')
plt.plot(x_vals, g(x_vals), linewidth=1, label='analytic derivative')
plt.scatter(x_vals, num_derivative, s=1, c='red', label = 'numerical derivative')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```
