## Fast Fourier Transforms


```python
import numpy as np 
from numpy.fft import rfft, irfft
```

```python
y=np.array([i/10 for i in range(10)], float)
c=rfft(y)
print(c)
```

```python
z=irfft(c)
print(z)
```

Fourier transforms of musical instruments

```python
from pylab import plot,xlim,show 
import matplotlib.pyplot as plt

piano=np.loadtxt('piano.txt', float)
trumpet=np.loadtxt('trumpet.txt', float)
x=[i for i in range(len(piano))]
plt.figure(dpi=200)
fig,axs=plt.subplots(2, figsize=(10,10))
axs[0].plot(x, piano, 'tab:green')
axs[0].set_title('Piano')
axs[1].plot(x, trumpet, 'tab:red')
axs[1].set_title('Trumpet')

for ax in axs.flat:
    ax.label_outer()
```

```python
piano_fft, trumpet_fft = rfft(piano), rfft(trumpet)
size=10000
fig,axs=plt.subplots(2, figsize=(10,10))
axs[0].plot(abs(piano_fft[:size]), 'tab:green')
axs[0].set_title('Piano')
axs[1].plot(abs(trumpet_fft[:size]), 'tab:red')
axs[1].set_title('Trumpet')
for ax in axs.flat:
    ax.label_outer()
```

```python
piano=np.loadtxt('piano.txt', float)
s, N= 44100, len(piano)
piano_fft = rfft(piano)
max_arg = abs(piano_fft[:10000]).argmax()
print(f'Note frequency: {max_arg/(N/s)} Hz')
```

To calculate the frequency,
$$f = \frac{k_{max}}{T}$$
where 
$$T = \frac{N}{\text{sampling rate}}$$

```python
print(f'Piano note frequency: {max_arg/(N/s)}')
print(f'Corresponding note: C5 (one octave higher than middle C)')
```
