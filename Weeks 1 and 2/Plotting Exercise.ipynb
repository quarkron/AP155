```python
## Create the objects
x = np.linspace(0, 4*np.pi, 100)
y = np.cos(x)
z = np.sin(x)
fig, ax = plt.subplots()

ax.plot(x, y, color='black', label="Signal A")
ax.plot(x, z, color='red', label="Signal B")

ax.set_xlabel(r"Time $t$") 
ax.set_ylabel(r"Output $V_{out}$")

ax.grid()
ax.legend(loc='best')

fig.tight_layout()
fig.savefig("siggen.png", dpi=150)
```
