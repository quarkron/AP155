```python
import numpy as np
from random import *
from numpy.random import *
import matplotlib.pyplot as plt
```

```python
num_atoms = 1000
initial_thal =  [1 for i in range(num_atoms/2)] 
lead = num_atoms/2
copy_thal = np.copy(initial_thal)

tau = 3.053*60
t = 0 
step = 1
prob = 1 - 2**(-step/tau)

while t < tau:
    for i in range(500): 
        if random() < prob and initial_thal[i] == 1:
            initial_thal[i] = 0
    t += 1
sum(initial_thal)
```

```python
num_atoms = 10000
lead = 0
thal = num_atoms


tau = 3.053*60
t = 0 
time = 20*tau
step = 1
decay_prob = 1 - 2**(-step/tau)

atoms_array = []
while t < time:
    for i in range(int(thal)): 
        if random() < decay_prob:
            thal -= 1
            lead += 1
    atoms_array.append([thal,lead])
    t += 1

a=np.array(atoms_array)
plt.plot(a[:,0])
plt.plot(a[:,1])
```

Faster implementation: Make use of exponential probability distribution

```python
tau = 3.053*60
num_atoms = 100000
scale = tau/np.log(2)

dist = np.random.exponential(scale=scale, size=num_atoms)
sorted=np.sort(dist)

x= np.arange(num_atoms)
plt.plot(sorted, x) 
plt.plot(sorted, num_atoms-x)

```
